```python
import matplotlib.pyplot as plt
growth = [5.80000327485702,4.72530605111095,5.04526758311702,7.04319246586394,11.4497432088167,9.40000151584914,9.99999473022963,8.60000659811426,10.1999994860937,0.199995231733979,7.75001707500896,5.54997471638077,1.70785253368828,0.998719299661161,1.65112470228097,-3.82957056812593,-2.5264464460425,2.53218349471909,3.14919758275296,1.44645548204305,-0.672742918585868,2.43886210234947,-4.65839591328999,4.12654007153881,4.00984852465452]
year = [n for n in range(2000, 2025)]

summa = 0
for num in growth:
    summa += num
avg = summa / len(growth)
```

```python
plt.plot(year, growth, label='Реальный прирост ВВП РБ, %', marker='s')
plt.xticks(year, rotation=45)
plt.axhline(0, linestyle='--', color = 'black')
plt.axhline(growth[-3], linestyle='--', color='red', label=f'Минимум: -4.7%')
plt.axhline(growth[4], linestyle='--', color='green', label=f'Максимум: 11.4%')
plt.axhline(avg, linestyle='--', label=f'Средний прирост: {round(avg, 2)}%')
plt.legend(loc=(0.45,0.7))

for x, y in zip(year, growth):
    plt.text(x, y+0.2, f'{y:.1f}',  
             ha='center', va='bottom', fontsize=8)

plt.show()
```
